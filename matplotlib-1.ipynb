#### Module 1 : Intro to Matplotlib

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
plt.plot([1,2,3],[4,5,6])
plt.title("Pyplo Example")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
# Object Oriented interface
fig , ax = plt.subplots()
ax.plot([1,2,3],[4,5,6])
ax.set_title("OO API Example")
ax.set_xlabel("X")
ax.set_ylabel("Y")
plt.show()
```

```python
# Configuration 
mpl.rcParams['font.size'] = 12
mpl.rcParams['lines.linewidth']=2
```

```python
# Example
fig = plt.figure(figsize=(6,4), dpi=120)
ax = fig.add_subplot(1,1,1)
x = [0,1,2,3,4]
y = [i**2 for i in x]
ax.plot(x,y ,marker='o',color='Teal',linestyle='--')
plt.show()
```

6. Customizing Plots

#### Line Styles & Colors

- `linestyle`: `'-'`, `'--'`, `':'`, `'-.'`
    
- `marker`: `'o'`, `'s'`, `'^'`, `'+'`, etc.
    
- `color`: named colors (`'red'`), hex (`'#FF0000'`), RGB tuples (`(1.0, 0.0, 0.0)`)
    
- Shorthand: `plt.plot(x, y, 'ro--')`  red circles with dashed line
    

#### Labels and Titles

- `plt.xlabel()`, `plt.ylabel()`, `plt.title()`
    
- Customize font size, weight, color via parameters
    

python

CopyEdit

`plt.xlabel('X axis', fontsize=12, fontweight='bold', color='navy')`

#### Legends

- Use `label='...'` in plotting command and `plt.legend()`
    
- Position via `plt.legend(loc='upper left')`
    

#### Grids

- `plt.grid(True, which='both', linestyle='--', linewidth=0.5)`
    
- Control grid type and axis
